```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
URL = 'https://www.amazon.com/s?k=apple+iphone+16&crid=3GE4DQGP3N01A&sprefix=apple+iphone%2Caps%2C293&ref=nb_sb_ss_ts-doa-p_3_12'
```

```python
HEADERS= ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/132.0.0.0 Safari/537.36',
          'Accept-Language':'en-US, en;q=0.5'}) 
```

```python
webpage= requests.get(URL, headers= HEADERS)
```

```python
soup = BeautifulSoup(webpage.content, "html.parser")
```

```python
soup
```

```python
links= soup.find_all("a",attrs={'class': "a-link-normal s-line-clamp-2 s-link-style a-text-normal"})
```

```python
links
```

```python
prod_link = links[0].get('href')
```

```python
product_list= "http://amazon.com" + prod_link
```

```python
product_list
```

```python
new_webpage = requests.get(product_list, headers= HEADERS)
```

```python
new_webpage
```

```python
new_soup = BeautifulSoup(new_webpage.content, "html.parser")
```

```python
new_soup
```

```python
new_soup.find("span", attrs={"id":"productTitle"}).text.strip()
```
